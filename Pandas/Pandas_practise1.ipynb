<h2 style="color:green">Different Ways of Creating Dataframe</h2>

```python
#way 1 (python dictionary list)
weather_data = {
'day': ['1/1/2017', '1/2/2017', '1/3/2017'],
'temperature': [32, 35, 28],
'windspeed': [6,7,2],
'event': ['Rain', 'Sunny', 'Snow']
}


df = pd.DataFrame(weather_data)

df
```

```python

# way 2 (python tuple list)
weather_data = [
('1/1/2017', 32, 6, 'Rain'),
('1/2/2017', 35, 7, 'Sunny'),
('1/3/2017', 28, 2, 'Snow')
]

df = pd.DataFrame(weather_data,columns=["day",'temp',"speed","event"])

```

```python
df
```

```python
# way 3 (python dictionary list)
weather_data = [
{'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
{'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
{'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'}
]

df = pd.DataFrame(weather_data)
```

```python
df
```

<h2 style="color:green">interpolate()</h2>

```python
df = df.infer_objects(copy=False)
df = df.interpolate(numeric_only=True)
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].interpolate()
```
