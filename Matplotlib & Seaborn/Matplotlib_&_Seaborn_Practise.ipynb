```python
insurance_data = pd.read_csv("insurance.csv")
```

```python
insurance_data
```

```python
plt.figure(figsize=(18,8))
sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'])
plt.show()
```

```python
sns.regplot(x=insurance_data['bmi'],y=insurance_data['charges'])
plt.show()
```

```python
sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'],hue=insurance_data['smoker'])
plt.show()
```

```python
sns.lmplot(x='bmi',y='charges',hue='smoker',data=insurance_data)
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.swarmplot(x=insurance_data['smoker'],y=insurance_data['charges'],size=2)
plt.show()
```

```python
iris_data = pd.read_csv("iris.csv",index_col="Id")
```

```python
plt.figure(figsize=(16,6))
sns.histplot(x=iris_data['Petal Length (cm)'])
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.histplot(iris_data,x=iris_data['Petal Length (cm)'],hue="Species")
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.kdeplot(x=iris_data['Petal Length (cm)'])
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.kdeplot(x=iris_data['Petal Length (cm)'],fill=True,hue=iris_data['Species'])
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.jointplot(x=iris_data['Petal Length (cm)'],y=iris_data['Petal Width (cm)'],hue=iris_data['Species'])
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.jointplot(x=iris_data['Petal Length (cm)'],y=iris_data['Petal Width (cm)'],kind="kde")
plt.show()
```
