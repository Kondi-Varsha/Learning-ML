# Airbnb NYC Listings — Data Cleaning

## Introduction

Real-world data is rarely clean. It comes with missing values, incorrect data types (like prices with $ signs), and inconsistencies. The Airbnb Open Data is a perfect example of this challenge.


This dataset contains Airbnb listings from New York City covering 
details like host information, location, room types, pricing, 
availability, and guest reviews across 26 columns.

---

## Approach

Before jumping into cleaning, we will first **understand the dataset**:
- What each column means
- Which columns are important for analysis
- What data quality issues exist

Only then we will proceed to **clean the data**.

---

## Objective
- Understand the raw Airbnb NYC dataset
- Identify and fix data quality issues such as:
  - Missing values
  - Incorrect data types
  - Duplicate records
  - Invalid values & outliers
  - Drop unnecessary columns

<h2 style="color:blue">Dropping Unnecessary Columns</h2>

```python
df.drop(columns=['host id','country','country code','license','house_rules'],inplace=True)
```

df.isnull().sum()

```python
df.dtypes
```

<h2 style="color:blue">Filling null values</h2>

```python
df.fillna(
    value={
        'name':'UNK',
        'host_identity_verified':'unconfirmed',
        'host name':'UNK host',
        'neighbourhood group':'UNK NHD GRP',
        'neighbourhood':'UNK NHD',
        'instant_bookable':df.instant_bookable.mode()[0],
        'availability 365':0,
        'minimum nights':0,
        'number of reviews':0,
        'reviews per month':0,
        'review rate number':0,
        'calculated host listings count':0,
        'availability 365':0,
        'Construction year':df['Construction year'].mode()[0],
        'cancellation_policy':df['cancellation_policy'].mode()[0],
        },inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(5)
```

<h2 style="color:blue">Formatting Price Columns</h2>

```python
df.price = df.price.str.replace('$',' ')
```

```python
df.price.head(3)
```

```python
df.price = df.price.str.replace(',','')
```

```python
df.price
```

```python
df.dtypes
```

```python
df['service fee'] = df['service fee'].str.replace('$','')
```

```python
df['service fee'].head(10)
```

<h2 style="color:blue">String Type Conversion</h2>

```python
cols = ['NAME','host name','neighbourhood group','neighbourhood','instant_bookable','cancellation_policy','room type'] 
df[cols]= df[cols].astype(str)
```

```python


df.dtypes
```

```python
type(df['NAME'][0])  # will show <class 'str'>
```

```python
# Clean and convert to numeric first
df['price'] = df['price'].astype(float)
df['service fee'] = df['service fee'].astype(float)

# Then fill NaN with group mean
df['price'] = df.groupby('room type')['price'].transform(lambda x: x.fillna(x.mean().round()))
df['service fee'] = df.groupby('room type')['service fee'].transform(lambda x: x.fillna(x.mean().round()))
```

```python
df.drop(columns = ['last review'],inplace=True)
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

<h2 style="color:blue">Handling Outliers</h2>

```python
print(df[df['minimum nights']>365].shape[0])
print(df[df['availability 365']>365].shape[0])
```

```python
#using clip() instead of completely removing them 
df['minimum nights'] = df['minimum nights'].clip(upper=365)
df['availability 365'] = df['availability 365'].clip(upper=365)
```

<h2 style="color:blue">Standardizing Text Columns</h2>

```python
df['neighbourhood group']=df['neighbourhood group'].str.replace('brookln','Brooklyn')
df['neighbourhood group']=df['neighbourhood group'].str.replace('manhatan','Manhattan')
```

```python
cols = ['neighbourhood group', 'neighbourhood', 'host_identity_verified','room type', 'cancellation_policy']
for col in cols:
    print(df[col].unique())
```

<h2 style="color:blue">Neighbourhood Analysis</h2>

```python
#Average Price by Neighbourhood Group

# Step 1 - calculate average 
avg_price = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False).reset_index()

avg_price = avg_price[avg_price['neighbourhood group'] != 'Unk Nhd Grp']

# Step 2 - plot it
ax = sns.barplot(data=avg_price, x='neighbourhood group', y='price')

# Add values on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='$%.0f', padding=3)
    
plt.title("Average Price by Neighbourhood Group",fontweight='bold')
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price ($)")
plt.show()
```

```python
# Which neighbourhood group has the highest average rating?

# Step 1 - Calculate average rating per neighbourhood group
avg_rating = df.groupby('neighbourhood group')['review rate number'].mean().sort_values(ascending=False).reset_index()


avg_rating = avg_rating[avg_rating['neighbourhood group'] != 'Unk Nhd Grp']

# Step 2 - Plot the bar chart
ax = sns.barplot(data=avg_rating, x='neighbourhood group', y='review rate number')

# Step 3 - Add values on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', padding=3)

# Step 4 - Add title and labels
plt.title("Average Rating by Neighbourhood Group", fontsize=14, fontweight='bold')
plt.xlabel("Neighbourhood Group", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.show()
```

```python
# Which neighbourhood is the most expensive?

# Step 1 - Calculate average price per neighbourhood
avg_price = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False).reset_index()

avg_price = avg_price[avg_price['neighbourhood group'] != 'Unk Nhd Grp']

# Step 2 - Plot the bar chart
ax = sns.barplot(data=avg_price, x='neighbourhood group', y='price')

# Step 3 - Add dollar values on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='$%.0f', padding=3)

# Step 4 - Add title and labels
plt.title("Most Expensive Neighbourhoods", fontsize=16, fontweight='bold')
plt.xlabel("Neighbourhood", fontsize=12)
plt.ylabel("Average Price ($)", fontsize=12)
plt.show()
```

```python
# Which neighbourhood group is least popular? (most available days)

avg_avail = df.groupby('neighbourhood group')['availability 365'].mean().sort_values(ascending=False).reset_index()

avg_avail = avg_avail[avg_avail['neighbourhood group'] != 'Unk Nhd Grp']


ax = sns.barplot(data=avg_avail, x='neighbourhood group', y='availability 365')

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3)

plt.title("Which Neighbourhood is Least Popular?", fontsize=16, fontweight='bold')
plt.xlabel("Neighbourhood Group", fontsize=12)
plt.ylabel("Avg Availability (days)", fontsize=12)
plt.show()
```

<h2 style="color:blue">Host Analysis</h2>

```python
# Which host has the most highly rated listings?

# Step 1 - Filter only high rated listings (rating 4 and above)
high_rated = df[df['review rate number'] >= 4]

# Fix - filter out unknown hosts
high_rated = high_rated[high_rated['host name'] != 'UNK host']

# Step 2 - Count listings per host
host_listings = high_rated.groupby('host name')['review rate number'].count().sort_values(ascending=False).reset_index().head(10)

# Step 3 - Rename column for clarity
host_listings.columns = ['host name', 'high rated listings count']

# Step 4 - Plot
ax = sns.barplot(data=host_listings, x='high rated listings count', y='host name')

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3)

plt.title("Top 10 Hosts with Most High Rated Listings", fontsize=16, fontweight='bold')
plt.xlabel("Number of High Rated Listings", fontsize=12)
plt.ylabel("Host Name", fontsize=12)
plt.tight_layout()
plt.show()
```

<h2 style="color:blue">Cancellation Policy Analysis</h2>

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1 - Higher reviews = more bookings
avg_reviews = df.groupby('cancellation policy')['number of reviews'].mean().sort_values(ascending=False).reset_index()
ax1 = sns.barplot(data=avg_reviews, x='cancellation policy', y='number of reviews', ax=axes[0])
for container in ax1.containers:
    ax1.bar_label(container, fmt='%.0f', padding=3)
axes[0].set_title("Cancellation Policy vs Reviews", fontsize=14, fontweight='bold')
axes[0].set_xlabel("Cancellation Policy")
axes[0].set_ylabel("Avg Number of Reviews")

# Plot 2 - Lower availability = more booked out
avg_avail = df.groupby('cancellation policy')['availability 365'].mean().sort_values(ascending=True).reset_index()
ax2 = sns.barplot(data=avg_avail, x='cancellation policy', y='availability 365', ax=axes[1])
for container in ax2.containers:
    ax2.bar_label(container, fmt='%.0f', padding=3)
axes[1].set_title("Cancellation Policy vs Availability", fontsize=14, fontweight='bold')
axes[1].set_xlabel("Cancellation Policy")
axes[1].set_ylabel("Avg Availability (days)")

plt.tight_layout()
plt.show()
```

```python
# Does flexible cancellation policy affect review rate?

# Step 1 - Calculate average review rate per cancellation policy
avg_review = df.groupby('cancellation policy')['review rate number'].mean().sort_values(ascending=False).reset_index()

# Step 2 - Plot the bar chart
ax = sns.barplot(data=avg_review, x='cancellation policy', y='review rate number')

# Step 3 - Add values on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', padding=3)

# Step 4 - Add title and labels
plt.title("Does Cancellation Policy Affect Review Rate?", fontsize=16, fontweight='bold')
plt.xlabel("Cancellation Policy", fontsize=12)
plt.ylabel("Average Review Rate", fontsize=12)
plt.show()
```

## Final Conclusion — Customer Perspective

###  Ideal Listing for Maximum Bookings
| Factor | Recommendation |
|--------|---------------|
| 📍 Location | Manhattan or Brooklyn |
| 🏠 Room Type | Entire Home or Private Room |
| 💰 Price | $50 - $300 per night |
| 📋 Cancellation | Flexible Policy |
| ⭐ Rating | All areas similar (3 - 3.5) |

###  Key Findings
- Price difference across neighbourhoods is only **$7** — room type drives price more than location.
- Entire Home/Apt is most expensive but also **most booked** — customers pay more for privacy.
- Flexible cancellation attracts **more bookings** but does NOT improve ratings.
- Staten Island and Bronx are **least popular** regardless of price.
- Building age has **no effect** on price or popularity.

###  Bottom Line
> A customer looking for the best NYC Airbnb experience would choose an **Entire Home or Private Room** in **Manhattan or Brooklyn**, priced **under $300**, with a **flexible cancellation policy**. Ratings are consistent everywhere so no need to stress about which neighbourhood to pick!
