# Airbnb NYC Listings — Data Cleaning & EDA

## Introduction

Real-world data is rarely clean. It comes with missing values, incorrect data types (like prices with $ signs), and inconsistencies. The Airbnb Open Data is a perfect example of this challenge.


This dataset contains Airbnb listings from New York City covering 
details like host information, location, room types, pricing, 
availability, and guest reviews across 25 columns.

---

## Approach

Before jumping into cleaning, we will first **understand the dataset**:
- What each column means
- Which columns are important for analysis
- What data quality issues exist

Only then we will proceed to **clean the data**.

---

## Objective
- Understand the raw Airbnb NYC dataset
- Identify and fix data quality issues such as:
  - Missing values
  - Incorrect data types
  - Duplicate records
  - Invalid values & outliers
  - Drop unnecessary columns

<h2 style="color:blue">Removal of Duplicates</h2>

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.columns
```

<h2 style="color:blue">Dropping Unnecessary Columns</h2>

```python
df.drop(columns=['host id','country','country code','license','house_rules'],inplace=True)
```

df.isnull().sum()

```python
df.dtypes
```

<h2 style="color:blue">Filling null values</h2>

```python
df.fillna(
    value={
        'name':'UNK',
        'host_identity_verified':'unconfirmed',
        'host name':'UNK host',
        'neighbourhood group':'UNK NHD GRP',
        'neighbourhood':'UNK NHD',
        'instant_bookable':df.instant_bookable.mode()[0],
        'availability 365':0,
        'minimum nights':0,
        'number of reviews':0,
        'reviews per month':0,
        'review rate number':0,
        'calculated host listings count':0,
        'availability 365':0,
        'Construction year':df['Construction year'].mode()[0],
        'cancellation_policy':df['cancellation_policy'].mode()[0],
        },inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(5)
```

<h2 style="color:blue">Formatting Price Columns</h2>

```python
df.price = df.price.str.replace('$',' ')
```

```python
df.price.head(3)
```

```python
df.price = df.price.str.replace(',','')
```

```python
df.price
```

```python
df.dtypes
```

```python
df['service fee'] = df['service fee'].str.replace('$','')
```

```python
df['service fee'].head(10)
```

<h2 style="color:blue">String Type Conversion</h2>

```python
cols = ['NAME','host name','neighbourhood group','neighbourhood','instant_bookable','cancellation_policy','room type'] 
df[cols]= df[cols].astype(str)
```

```python
df.dtypes
```

```python
type(df['NAME'][0])  # will show <class 'str'>
```

```python
# Clean and convert to numeric first
df['price'] = df['price'].astype(float)
df['service fee'] = df['service fee'].astype(float)

# Then fill NaN with group mean
df['price'] = df.groupby('room type')['price'].transform(lambda x: x.fillna(x.mean().round()))
df['service fee'] = df.groupby('room type')['service fee'].transform(lambda x: x.fillna(x.mean().round()))
```

```python
df.drop(columns = ['last review'],inplace=True)
```

```python
df.head(2)
```

<h2 style="color:blue">Removed null values</h2>

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.max()
```

<h2 style="color:blue">Handling Outliers</h2>

```python
print(df[df['minimum nights']>365].shape[0])
print(df[df['availability 365']>365].shape[0])
```

```python
#using clip() instead of completely removing them 
df['minimum nights'] = df['minimum nights'].clip(upper=365)
df['availability 365'] = df['availability 365'].clip(upper=365)
```

```python
print(df[df['availability 365']>365].shape[0])
print(df[df['minimum nights']>365].shape[0])
```

```python
df.dtypes
```

```python
df.max()
```

<h2 style="color:blue">Checking Unique Values</h2>

```python
print(df['neighbourhood group'].unique())
print(df['neighbourhood'].unique())
print(df['host_identity_verified'].unique())
print(df['cancellation_policy'].unique())
print(df['room type'].unique())
```

<h2 style="color:blue">Standardizing Text Columns</h2>

```python
df['neighbourhood group']=df['neighbourhood group'].str.replace('brookln','Brooklyn')
df['neighbourhood group']=df['neighbourhood group'].str.replace('manhatan','Manhattan')
```

<h2 style="color:blue">Capitalizing Column Values</h2>

```python
cols = ['neighbourhood group', 'neighbourhood', 'host_identity_verified','room type', 'cancellation_policy']
for col in cols:
    df[col] = df[col].str.strip().str.title()
```

```python
cols = ['neighbourhood group', 'neighbourhood', 'host_identity_verified','room type', 'cancellation_policy']
for col in cols:
    print(df[col].unique())
```

<h2 style="color:blue">Renaming Column Names</h2>

```python
df.rename(columns={'NAME':'name'},inplace=True)
```

```python
df.head(2)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace('_'," ")
```

<h2 style="color:blue">Converting Float to Int</h2>

```python
cols = ['construction year','price','service fee','minimum nights','number of reviews','reviews per month','review rate number','calculated host listings count','availability 365']
for col in cols:
    df[col]=df[col].astype(int)
```
